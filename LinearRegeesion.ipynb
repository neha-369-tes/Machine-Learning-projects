```python
import pandas as pd
import numpy as np

np.random.seed(42)

YearsExperience = np.random.randint(0, 11, 30)
Salary = 30000 + YearsExperience * 5000 + np.random.randint(-4000, 4000, 30)

df = pd.DataFrame({
    "YearsExperience": YearsExperience,
    "Salary": Salary
})

print(df.head())
```

```python
def gradient_descent(m_now, b_now, points, L):
  m_gradient = 0
  b_gradient = 0
  n = len(points)
  for i in range(n):
    x= points.iloc[i].YearsExperience
    y = points.iloc[i].Salary
    m_gradient += -(2/n) * x * (y -(m_now * x + b_now))
    b_gradient += -(2/n) *(y - (m_now * x + b_now))
  m = m_now - m_gradient * L
  b = b_now - b_gradient * L
  return m , b
m = 0
b = 0
L = 0.001
epochs = 1000
for i in range(epochs):
  m, b = gradient_descent(m, b, df, L)
print(m, b)
x_vals = np.array(df.YearsExperience)
plt.scatter(df.YearsExperience, df.Salary, color="black")
plt.plot(x_vals, m * x_vals + b, color="red")
plt.show()
```
